```python
import pandas as pd
import re
```

```python
df = pd.read_json("../scraping/j_raw_deals_2.json", convert_dates=False)
```

```python
df
```

```python
def wo_remove(text):
    remove = "wo"
    remove_2 = "di"
    text = text.replace(remove, "")
    text = text.replace(remove_2, "").strip()
    return text

df["date"] = df["date"].apply(wo_remove)
df
```

```python
df["date"] = df["date"].str.split("t/m")

def split_date(date_val):

    if isinstance(date_val, list):
        cleaned = ' '.join(str(x) for x in date_val)
    else:
        cleaned = str(date_val)

    cleaned = cleaned.replace('\xa0', ' ').replace('NBSP', '').strip()
    cleaned = re.sub(r'\s+', ' ', cleaned).strip()

    same_month = re.match(r'(\d+)\s+(\d+)\s+(\w+)', cleaned)
    if same_month:
        day1, day2, month = same_month.groups()
        return f"{day1} {month}", f"{day2} {month}"

    diff_month = re.match(r'(\d+)\s+(\w+)\s+(\d+)\s+(\w+)', cleaned)
    if diff_month:
        day1, month1, day2, month2 = diff_month.groups()
        return f"{day1} {month1}", f"{day2} {month2}"

    return cleaned, cleaned

df[['start_date', 'end_date']] = df['date'].apply(
    lambda x: pd.Series(split_date(x))
)
```

```python
df
```

```python
df = df.drop("date", axis=1)
df
```

```python
def product_clean(text):
    word = "[Kompas JumRouterLink]"
    text = text.replace(word, "").strip()
    return text

df["product"] = df["product"].apply(product_clean)
df
```

```python
#removing products which are online or liquor store
df = df[~df["discount"].str.contains("Alleen online", case=True, na=False)]
df = df[~df["discount"].str.contains("slijterij", case=True, na=False)]
df = df[df['product'] != 'Welke biermerken zijn bij jouw Jumbo in de aanbieding?']
```

```python
df
```

```python
def winkel(text):
    wink = "Alleen in de winkel"
    if wink in text:
        return text.replace(wink, "").strip()
    return text

df["discount"] = df["discount"].apply(winkel)
df
```

```python
df.to_json('../../data/J_cleaned_7.json', orient='records', indent=4, force_ascii=False)
```
