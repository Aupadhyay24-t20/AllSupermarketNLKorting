```python
import pandas as pd
```

```python
df = pd.read_json("../scraping/new_raw_deals.json")
```

```python
df
```

```python
df["date"] = df["date"].str.replace(" t/m ", " ")
def date_fixer(df):
    patern = r'\d+\s\w+'
    df["start_date"] = df["date"].str.findall(patern).str[0]
    df["end_date"] = df["date"].str.findall(patern).str[1]
    return df

df = date_fixer(df)
df = df.drop("date", axis = 1)



```

```python
df["discount"] = df["discount"].str.split(":")
```

```python
df["discount"] = df["discount"].str[1]
```

```python
df["discount"]
```

```python
df["discount"] = df["discount"].str.split(",", n=1)
df["discount"] = df["discount"].str[1]
df["discount"]

```

```python
#Deleting all rows which have either bezorging or online.
df = df[~df["product"].str.contains("bezorging", case=False, na=False)]
```

```python
df = df[~df["product"].str.contains("Online", case=True, na=False)]
df
```

```python
#Removing all products which are online
target_index = df[df["product"].str.contains("OP=OP", case=True, na=False)].index[0]
df = df.loc[:target_index - 1]
df
```

```python
#cleaning the remaining discount values
def discount_clean(df):
    df["product"] = df["product"].str.split(", ", expand=False)
    df["discount"] = df["discount"].str.split(", ", expand=False)
    mask = df["product"].str.len() > 1
    df.loc[mask, "discount"] = df.loc[mask, "discount"].str[1]
    return df

df_final = df.copy()
df_final = discount_clean(df_final)
df_final["product"] = df_final["product"].str.join(', ')
df_final["discount"] = df_final["discount"].apply(
    lambda x: ', '.join(x) if isinstance(x, list) else x
)
df_final
```

```python
df_final.to_json('data/AH_cleaned.json', orient='records', indent = 4, force_ascii=False)
```
