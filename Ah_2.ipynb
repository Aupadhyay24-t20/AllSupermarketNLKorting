```python
import pandas as pd
```

```python
df = pd.read_json("new_raw_deals.json")
```

```python
df
```

```python
df["date"] = df["date"].str.replace(" t/m ", " ")
df["date"]
def date_fixer(df):
    date = df["date"].str.split(expand=True)
    start = date[0]
    end = date[1]
    month = date[2]
    start_date = start + " " + month
    end_date = end + " " + month
    df["start_date"] = start_date
    df["end_date"] = end_date
    return df

df = date_fixer(df)
df = df.drop("date", axis=1)
df
```

```python
df["discount"] = df["discount"].str[0]
df
```

```python
df["discount"] = df["discount"].str.split(":")
df
```

```python
df["discount"] = df["discount"].str[1]
df
```

```python
#Deleting all rows which have bezorging.
df = df[~df["product"].str.contains("bezorging", case=False, na=False)]
df
```

```python
df = df[~df["product"].str.contains("Online", case=True, na=False)]
df
```

```python
#Removing all products which are online
target_index = df[df["product"].str.contains("OP=OP", case=True, na=False)].index[0]
df = df.loc[:target_index - 1]
df
```

```python
import re

def discount_clean(text):
    pattern = r'(van \d*\.?\d* voor \d*\.?\d*)'
    match = re.search(pattern, text)
    if match:
        return match.group(1)
    else:
        return text
df["discount"] = df["discount"].apply(discount_clean)
df
```

```python
#handling the scenarios with voor only prices
def voor_filter(text):
    pattern = r'((?:\d+\W*?\w*\s+)?voor \d+\.?\d*)'
    text_van = r'(van \d*\.?\d* voor \d*\.?\d*)'
    text_pattern = re.search(text_van, text)
    if text_pattern:
        return text
    else:
        match = re.search(pattern, text)
        if match:
            return match.group(1)
        else:
            return text

df["discount"] = df["discount"].apply(voor_filter)
df
```

```python
def discount_name_cleaner(row):
    discount = str(row["discount"])
    product = str(row["product"])
    if(product in discount):
        return discount.replace(product, "").strip()
    return discount

df_new = df.copy()
df_new["discount"] = df_new.apply(discount_name_cleaner, axis=1)
df_new
```

```python
#removing Allen in winkel
def winkel(text):
    wink = "Alleen in de winkel"
    if wink in text:
        return text.replace(wink, "").strip()
    return text
df_new["discount"] = df_new["discount"].apply(winkel)
df_new
```

```python
df_new.to_json('data/AH_cleaned_3.json', orient='records', indent=4, force_ascii=False)
```
